```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import numpy as np
from sklearn.datasets import fetch_mldata
mnist= fetch_mldata('MNIST original')
```

```python
X, y = mnist["data"], mnist["target"]
X.shape
```

```python
some_digit= X[67000]
some_digit_image= some_digit.reshape(28, 28)
plt.imshow(some_digit_image, cmap= matplotlib.cm.binary,
interpolation="nearest")
plt.axis("off")
plt.show()
```

```python
X_train, X_test, y_train, y_test= X[:60000], X[60000:], y[:60000], y[60000:]
```

```python
shuffle_index= np.random.permutation(60000)
X_train, y_train= X_train[shuffle_index], y_train[shuffle_index]
```

```python
pca.explained_variance_ratio_
```

```python
pca= PCA(n_components= 154)
X_reduced= pca.fit_transform(X_train)
X_recovered= pca.inverse_transform(X_reduced)
pca.components_.shape
```

```python
fig , ax = plt.subplots(2,10,figsize=(20,5), subplot_kw={'xticks':[], 'yticks':[]}, gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i in range(10):
    ax[0,i].imshow(X_train[12000+i].reshape(28,28), cmap= 'binary_r', interpolation="nearest")
    ax[1,i].imshow(X_recovered[12000+i].reshape(28,28), cmap= 'binary_r', interpolation="nearest")
ax[0,0].set_ylabel('Full dimention')    
ax[1,0].set_ylabel('Reconstructed Figures')
```

```python
fig , axes = plt.subplots(2,10,figsize=(20,5), subplot_kw={'xticks':[], 'yticks':[]}, gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i,axe in enumerate(axes.flat):
    axe.imshow(pca.components_[i].reshape(28,28), cmap='bone')
```
